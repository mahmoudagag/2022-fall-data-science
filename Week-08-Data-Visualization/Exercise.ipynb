```python
df = pd.read_csv("data/Sales.csv")
df.head()
```

```python
items = df.columns[1:]
sns.lineplot(data=df[items])
```

```python
print("It tells us the volume being sold of each item")
print('I would recommend pushing bathing soap and would stop selling moisturizer')
```

```python
total_sales = df[items].sum().sum()
for item in items:
    print(f"{item} percentage:\t {round(((df[item].sum() / total_sales) * 100), 2)}%")
```
